```python
import pandas as pd

# Load the data
info_df = pd.read_csv('info.csv')

# Split "Hero|Publisher" into two separate columns
info_df[['Hero', 'Publisher']] = info_df['Hero|Publisher'].str.split('|', expand=True)
info_df.drop('Hero|Publisher', axis=1, inplace=True)

# Extract Height and Weight from Measurements
info_df['Height'] = info_df['Measurements'].apply(lambda x: float(eval(x)['Height'].split()[0]))
info_df['Weight'] = info_df['Measurements'].apply(lambda x: float(eval(x)['Weight'].split()[0]))

# Drop the Measurements column
info_df.drop('Measurements', axis=1, inplace=True)

# Load powers data
powers_df = pd.read_csv('powers.csv')

# Split Powers into a list
powers_df['Powers'] = powers_df['Powers'].str.split(',')

# Perform one-hot encoding on powers
powers_df = powers_df.join(pd.get_dummies(powers_df['Powers'].apply(pd.Series).stack()).sum(level=0))

# Merge the two DataFrames
df = pd.merge(info_df, powers_df, left_on='Hero', right_on='hero_names')

# Drop the redundant hero_names column
df.drop('hero_names', axis=1, inplace=True)

# Print the DataFrame
print(df)

# Answer the questions
avg_weight_super_speed = df[df['Super Speed'] == 1]['Weight'].mean()
avg_weight_no_super_speed = df[df['Super Speed'] == 0]['Weight'].mean()
print("Average weight of heroes with Super Speed: ", avg_weight_super_speed)
print("Average weight of heroes without Super Speed: ", avg_weight_no_super_speed)

avg_height_publisher = df.groupby('Publisher')['Height'].mean()
print("Average height of heroes for each publisher: ", avg_height_publisher)
```
